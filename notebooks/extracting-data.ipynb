```python
# ============================================================
# Step 1 — Load only Shot events from StatsBomb JSON files
# ============================================================

import json
import glob
from pathlib import Path

import numpy as np
import pandas as pd

# Your StatsBomb events folder
events_folder = "/Users/farzammanafzadeh/Desktop/Module 3 FAKIR Project/StatsBomb/statsbomb-events/data/events"

# Find all JSON event files
json_files = sorted(glob.glob(events_folder + "/*.json"))

print("Number of JSON files found:", len(json_files))

all_shots = []

for file in json_files:
    match_id = Path(file).stem

    with open(file, "r", encoding="utf-8") as f:
        events = json.load(f)

    # Keep only Shot events from this match
    shot_events = [
        event for event in events
        if isinstance(event.get("type"), dict)
        and event["type"].get("name") == "Shot"
    ]

    # Add match_id to each shot
    for shot in shot_events:
        shot["match_id"] = match_id

    all_shots.extend(shot_events)

print("Total shot events found:", len(all_shots))

shots_raw = pd.DataFrame(all_shots)

print("Raw shots dataframe shape:", shots_raw.shape)

display(shots_raw.head())
```

```python
# ============================================================
# Step 2 FIXED — Flatten raw StatsBomb shot dataframe
# Handles both real dictionaries and string dictionaries
# Raw dataframe name: df
# Output dataframe name: shots_df
# ============================================================

import ast
import numpy as np
import pandas as pd


def parse_if_string(value):
    """
    Convert string representations of dictionaries/lists back to Python objects.
    If the value is already a dict/list, keep it unchanged.
    """
    if isinstance(value, (dict, list)):
        return value
    
    if pd.isna(value):
        return np.nan
    
    if isinstance(value, str):
        try:
            return ast.literal_eval(value)
        except Exception:
            return value
    
    return value


def get_name(value):
    """
    Extract the 'name' field from a StatsBomb dictionary.
    Works even if the dictionary is stored as a string.
    """
    value = parse_if_string(value)
    
    if isinstance(value, dict):
        return value.get("name")
    
    return np.nan


def get_location_x(location):
    """
    Extract x coordinate from location [x, y].
    Works even if location is stored as a string.
    """
    location = parse_if_string(location)
    
    if isinstance(location, list) and len(location) >= 2:
        return location[0]
    
    return np.nan


def get_location_y(location):
    """
    Extract y coordinate from location [x, y].
    Works even if location is stored as a string.
    """
    location = parse_if_string(location)
    
    if isinstance(location, list) and len(location) >= 2:
        return location[1]
    
    return np.nan


def get_shot_value(shot_dict, key):
    """
    Extract direct value from the nested shot dictionary.
    Example: shot['statsbomb_xg']
    """
    shot_dict = parse_if_string(shot_dict)
    
    if isinstance(shot_dict, dict):
        return shot_dict.get(key)
    
    return np.nan


def get_shot_name(shot_dict, key):
    """
    Extract shot[key]['name'] from nested shot dictionary.
    Example: shot['outcome']['name']
    """
    shot_dict = parse_if_string(shot_dict)
    
    if isinstance(shot_dict, dict):
        value = shot_dict.get(key)
        value = parse_if_string(value)
        
        if isinstance(value, dict):
            return value.get("name")
    
    return np.nan


# Create clean shot-level dataframe
shots_df = pd.DataFrame()

# Event identifiers
shots_df["shot_id"] = df["id"]
shots_df["match_id"] = df["match_id"]

# Match time
shots_df["period"] = df["period"]
shots_df["minute"] = df["minute"]
shots_df["second"] = df["second"]

# Team and tactical context
shots_df["team"] = df["team"].apply(get_name)
shots_df["possession_team"] = df["possession_team"].apply(get_name)
shots_df["play_pattern"] = df["play_pattern"].apply(get_name)
shots_df["position"] = df["position"].apply(get_name)

# Shot coordinates
shots_df["x"] = df["location"].apply(get_location_x)
shots_df["y"] = df["location"].apply(get_location_y)

# Shot-specific nested features
shots_df["statsbomb_xg"] = df["shot"].apply(lambda s: get_shot_value(s, "statsbomb_xg"))
shots_df["shot_outcome"] = df["shot"].apply(lambda s: get_shot_name(s, "outcome"))
shots_df["body_part"] = df["shot"].apply(lambda s: get_shot_name(s, "body_part"))
shots_df["technique"] = df["shot"].apply(lambda s: get_shot_name(s, "technique"))
shots_df["shot_type"] = df["shot"].apply(lambda s: get_shot_name(s, "type"))

# Boolean shot features
shots_df["first_time"] = df["shot"].apply(
    lambda s: bool(get_shot_value(s, "first_time")) if not pd.isna(get_shot_value(s, "first_time")) else False
)

shots_df["one_on_one"] = df["shot"].apply(
    lambda s: bool(get_shot_value(s, "one_on_one")) if not pd.isna(get_shot_value(s, "one_on_one")) else False
)

shots_df["open_goal"] = df["shot"].apply(
    lambda s: bool(get_shot_value(s, "open_goal")) if not pd.isna(get_shot_value(s, "open_goal")) else False
)

# Freeze-frame kept for later advanced feature engineering
shots_df["freeze_frame"] = df["shot"].apply(lambda s: get_shot_value(s, "freeze_frame"))

# Under pressure
shots_df["under_pressure"] = df["under_pressure"].fillna(False).astype(bool)

# Target variable
shots_df["goal"] = (shots_df["shot_outcome"] == "Goal").astype(int)

# Convert numeric columns
numeric_columns = ["x", "y", "statsbomb_xg", "minute", "second"]

for col in numeric_columns:
    shots_df[col] = pd.to_numeric(shots_df[col], errors="coerce")

print("Clean shot dataframe shape:", shots_df.shape)

display(shots_df.head())
```
